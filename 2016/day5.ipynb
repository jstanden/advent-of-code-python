```python
from hashlib import md5
from itertools import count
```

```python
with open("../data/2016/day5.txt") as f:
    door_id = f.read().strip()
```

```python
def part1(s: str, length: int):
    password, counter, last_i = '', count(-1), 0
    while length:
        if (h := md5(f"{s}{next(counter)}".encode()).hexdigest()).startswith('00000'):
            password += h[5]
            length -= 1
    return password
```

```python
print(f"Part 1: {part1(door_id, 8)}")
```

```python
def part2(s: str, length: int):
    password, counter = [None] * length, count(-1)
    while not all(password):
        if (h := md5(f"{s}{next(counter)}".encode()).hexdigest()).startswith('00000'):
            at, char = int(h[5]) if h[5].isdigit() else None, h[6]
            if at is not None and -1 < at < length and password[at] is None:
                password[at] = char
    return "".join(password)

print("Part 2:", part2(door_id, 8))
```
