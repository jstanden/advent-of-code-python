```python
from collections import Counter
```

```python
with open("../data/2016/day6.txt") as f:
    lines = f.read().splitlines()
```

```python
part1, part2 = '', ''
for i in range(len(lines[0])):
    counts = Counter([line[i] for line in lines]).most_common()
    part1 += counts[0][0]
    part2 += counts[-1][0]
print(f"Part 1: {part1}")
print(f"Part 2: {part2}")
```
