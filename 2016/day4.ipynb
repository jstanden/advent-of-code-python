```python
from collections import Counter
from itertools import islice
```

```python
with open("../data/2016/day4.txt") as f:
    lines = f.read().splitlines()
```

```python
invert_letter = str.maketrans((alpha := "".join([chr(d) for d in range(97, 123)])), alpha[::-1])

def top_letters(room: str):
    return "".join([l for l,_ in list(islice(sorted(Counter(room.replace('-','')).items(),
       key=lambda v: (v[1],v[0].translate(invert_letter)),  # break ties alphabetically
       reverse=True
    ), 5))])

rooms = [[room, top_letters(room), *sector.rstrip(']').split('[')]
    for room, sector in [line.rsplit('-', 1) for line in lines]]

valid_rooms = list(filter(lambda room: room[1] == room[3], rooms))
```

```python
print("Part 1: {}".format(sum([int(sector) for _ ,_, sector, _ in valid_rooms])))
```

```python
def rotate_letter(l: str, by: int):
    return ' ' if l == '-' else chr(97 + ((ord(l) - 97 + by) % 26))

for room in rooms:
    decrypted = "".join([rotate_letter(l, int(room[2])) for l in list(room[0])])
    if decrypted == "northpole object storage":
        print(f"Part 2: {room[2]}"); break
```
