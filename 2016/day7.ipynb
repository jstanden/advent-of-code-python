```python
import re
from itertools import chain
```

```python
with open("../data/2016/day7.txt") as f:
    lines = f.read().splitlines()
```

```python
ips = [re.split(r'(\[[a-z]+\])', line) for line in lines]
```

```python
def supports_tls(text: str) -> int:
    is_hypernet = text.startswith('[')
    match = re.search(r'(.)((?!\1).)\2\1', text.strip('[]'))   # Find distinct mirrored sequences
    if is_hypernet and match is not None: return -100      # Score below zero for invalid hypernet
    elif not is_hypernet and match is not None: return 1   # Score 1 for each valid supernet
    else: return 0   # ignore everything else

def supports_ssl(sequences: list[str]) -> bool:
    supernets = [seq for seq in sequences if not seq.startswith('[')]
    hypernets = [seq.strip('[]') for seq in sequences if seq.startswith('[')]
    aba_pattern = r'(?=((.)((?!\2).)\2))'
    abas = set(chain(*[[m.group(1) for m in re.finditer(aba_pattern, net)] for net in supernets]))
    babs = [bab for bab in map(lambda aba: aba[1]+aba[0]+aba[1], abas) if any([bab in net for net in hypernets])]
    return len(abas) > 0 and len(babs) > 0
```

```python
part1 = sum([1 if sum([supports_tls(seq) for seq in ip]) > 0 else 0 for ip in ips])
print(f"Part 1: {part1}")
```

```python
part2 = sum([1 if supports_ssl(ip) else 0 for ip in ips])
print(f"Part 2: {part2}")
```
