```python
import re
import numpy as np
```

```python
with open('../data/2016/day3.txt', 'r') as f:
    data = f.read()
```

```python
triangles = [tuple(map(int,re.match(r'(\d+)\s+(\d+)\s+(\d+)', line.strip()).groups()))
             for line in data.splitlines()]
```

```python
# Part 1
part1 = 0
for x, y, z in triangles:
    if x + y > z and y + z > x and z + x > y: part1 += 1
print("Part 1:", part1)
```

```python
# Part 2
part2 = 0
for x, y, z in np.array(triangles).reshape(-1, order='F').reshape(-1, 3):
    if x + y > z and y + z > x and z + x > y: part2 += 1
print("Part 2:", part2)
```
