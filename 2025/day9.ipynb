```python
from itertools import combinations
from shapely import Polygon
```

```python
with open("../data/2025/day9.txt") as f:
    data = f.read()
```

```python
coords = [tuple(map(int, line.split(','))) for line in data.splitlines()]
```

```python
rects = [[ax, ay, bx, by, (abs(ax-bx)+1)*(abs(ay-by)+1)]
    for (ax,ay), (bx,by) in combinations(coords, 2)]
rects = sorted(rects, key=lambda r: r[4], reverse=True)
print("Part 1:", rects[0][4])
```

```python
p = Polygon(coords + [coords[0]])
for ax, ay, bx, by, area in rects:
    if Polygon([(ax, ay), (ax, by), (bx, by), (bx, ay), (ax, ay)]).within(p):
        print("Part 2:", area); break
```
