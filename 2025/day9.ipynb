```python
from collections import defaultdict
from itertools import combinations
```

```python
with open("../data/2025/day9.txt") as f:
    data = f.read()
```

```python
grid = defaultdict(None)
coords = [tuple(map(int, line.split(','))) for line in data.splitlines()]
rects = [[ax, ay, bx, by, abs(ax-bx+1)*abs(ay-by+1)] for (ax,ay), (bx,by) in combinations(coords, 2)]
print("Part 1:", sorted(rects, key=lambda r: r[4], reverse=True)[0][4])
```
