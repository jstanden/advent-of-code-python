```python
import re
from math import prod
```

```python
with open("../data/2025/day6.txt") as f:
    data = f.read()
```

```python
# Condense to single spaces, split, and trim
instructions = list(map(lambda line: line.strip().split(' '), re.sub(' +', ' ', data).splitlines()))
operators = instructions.pop()  # The last line has operators
subtotals = [0 if o == '+' else 1 for o in operators]  # Init 0 for addition, 1 for multiplication

for row in instructions:
    for col_num, n in enumerate(row):
        if operators[col_num] == '+': subtotals[col_num] += int(n)
        else: subtotals[col_num] *= int(n)
print(f"Part 1: {sum(subtotals)}")
```

```python
instructions = data.splitlines()
operators = list(re.finditer(r'(\S +)', instructions.pop()))   # positions of all operators with column widths
instructions = [[i[m.start():m.end()] for i in instructions] for m in operators]   # numbers in columns not rows
```

```python
part2 = 0
for col_num, col in enumerate(instructions):
    offsets = range(1, len(col[0])+1)   # extract digits from the right
    # Assemble digits in the set by index, ignore blanks, remove empty strings, convert to ints
    numbers = map(int, filter(None, ["".join([num[-o] for num in col if num[-o] != ' ']) for o in offsets]))
    # ['328', '64 ', '98 '] -> + [8, 248, 369] = 625
    part2 += prod(numbers) if operators[col_num][0][0] == '*' else sum(numbers)
print(f"Part 2: {part2}")
```
