```python
with open("../data/2025/day1.txt") as f:
    lines = f.read()
```

```python
dial = 50
part1, part2 = 0, 0
```

```python
for direction, clicks in map(lambda line: (-1 if 'L' == line[0] else 1, int(line[1:])), lines.splitlines()):
    delta = direction * (clicks % 100)
    was_dial, dial = dial, dial + delta
    part2 += ((clicks // 100)  # full rotations
        # plus rotations landing on zero, or crossing zero excluding starting at zero
        + (1 if ((0 == dial) or (was_dial and dial != dial % 100)) else 0))
    dial = dial % 100 # constrain to 0-99 bounds
    if 0 == dial: part1 += 1
```

```python
print(f"Part 1: {part1}")
print(f"Part 2: {part2}")
```
