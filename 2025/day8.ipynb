```python
import networkx as nx
from itertools import combinations, count
from math import sqrt, prod
from heapq import heappush, heappop
```

```python
with open("../data/2025/day8.txt") as f:
    lines = f.read().splitlines()
```

```python
G = nx.Graph()
coords = {line: list(map(int, line.split(','))) for line in lines}
for c_id in coords.keys(): G.add_node(c_id)
```

```python
dists = []
part1, part2 = 0, 0
iterations, MAX_ITERATIONS = count(), 1000

def dist(v1, v2):
    return sqrt((v1[0] - v2[0]) ** 2 + (v1[1] - v2[1]) ** 2 + (v1[2] - v2[2]) ** 2)

for (a_id,a),(b_id,b) in combinations(coords.items(), 2):
    heappush(dists, (dist(a, b), a_id, b_id))

while True:
    dist, a_id, b_id = heappop(dists)
    G.add_edge(a_id, b_id)
    circuits = list(nx.connected_components(G))
    if next(iterations) == MAX_ITERATIONS-1:
        part1 = prod(sorted([len(g) for g in circuits], reverse=True)[:3])
    if 1 == len(circuits):
        part2 = coords[a_id][0] * coords[b_id][0]; break

print(f"Part 1: {part1}")
print(f"Part 2: {part2}")
```
