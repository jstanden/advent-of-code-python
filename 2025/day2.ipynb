```python
with open("../data/2025/day2.txt") as f:
    lines = f.read()
```

```python
part1, part2 = 0, 0
```

```python
# Repeating substrings
def has_repeat(s: str):
    # Duplicate the string and remove the first/last character to match substrings
    idx = (s + s)[1:-1].find(s)
    # We have a match if the pattern reappears
    return int(s[:idx+1]) if idx != -1 else None
```

```python
# Loop through ID ranges
for lo, hi in map(lambda line: map(int, line.split('-')), lines.split(',')):
    part1 += sum(filter(
        # Even-length numbers where both halves match
        lambda n: 0 == (digits := len(str(n))) % 2 and str(n)[0:digits//2] == str(n)[digits//2:],
        range(lo, hi+1)
    ))
    part2 += sum(filter(
        lambda n: has_repeat(str(n)),
        range(lo, hi+1)
    ))
```

```python
print(f"Part 1: {part1}")
print(f"Part 2: {part2}")
```
