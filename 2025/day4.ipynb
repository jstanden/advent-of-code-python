```python
from collections import defaultdict
```

```python
with open("../data/2025/day4.txt") as f:
    data = f.read()
```

```python
grid = defaultdict(None)
counts, accessible = [], set()
border = (-1+1j, 1j, 1+1j, -1, 1, -1-1j, -1j, 1-1j) # moore neighborhood
```

```python
# build grid; reverse the lines so grid "up" is +1j, key on complex coords
for y, line in enumerate(reversed(data.splitlines())):
    for x, tile in enumerate(line): grid[x + y * 1j] = tile
```

```python
while None == accessible.clear():
    for xy, tile in {xy: tile for xy, tile in grid.items() if tile == '@'}.items(): # find all '@'
        if 4 > len([neighbor for offset in border if (neighbor := grid.get(xy + offset)) == '@']):
            accessible.add(xy) # add to list if less than 4 '@' neighbors
    counts.append(len(accessible))
    if 0 == len(accessible): break # stop when we run out of accessible paper rolls
    else: [grid.pop(xy) for xy in accessible] # remove accessible rolls to unblock others
```

```python
print(f"Part 1: {counts[0]}")
print(f"Part 2: {sum(counts)}")
```
