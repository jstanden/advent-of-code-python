```python
from functools import cmp_to_key
```

```python
with open("../data/2025/day5.txt") as f:
    data = f.read()
```

```python
def inclusive_range(start, end, step=1): return range(start, end+1, step)
fresh_ranges, ingredients = map(lambda line: line.splitlines(), data.split("\n\n"))
fresh_ranges = list(map(lambda r: inclusive_range(*map(int, r.split('-'))), fresh_ranges))
```

```python
part1 = 0
for ingredient in map(int, ingredients):
    part1 += 1 if any(ingredient in r for r in fresh_ranges) else 0
print(f"Part 1: {part1}")
```

```python
def sort_ranges_by_start (a, b): return (a.start > b.start) - (a.start < b.start)

def merge_ranges(ranges: list) -> list:
    # Sort ranges by starting index and propagate merge forward
    i, ranges = 0, sorted(set(ranges), key=cmp_to_key(sort_ranges_by_start))
    while i+1 < len(ranges): # while we have a next range
        if ranges[i + 1].start >= ranges[i].start and ranges[i + 1].stop <= ranges[i].stop: # next within current
            del ranges[i+1] # remove the next range
            i -= 1 # repeat this iteration with our new range
        elif ranges[i + 1].start < ranges[i].stop <= ranges[i + 1].stop: # next right overlaps current
            ranges[i] = range(ranges[i].start, ranges[i + 1].stop) # extend current stop
            i -= 1 # repeat this iteration with our new range
        i += 1
    return ranges

print("Part 2: " + str(sum([r.stop-r.start for r in merge_ranges(fresh_ranges)])))
```
