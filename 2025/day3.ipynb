```python
with open("../data/2025/day3.txt") as f:
    data = f.read()
```

```python
def highest_in_segment(segment: str, min_length: int):
    # Ignore digits at the end if we'd run out of digits
    until = -min_length + 1 if min_length > 1 else None
    # Find the next highest digit with a sufficient remainder
    index = segment.find(max(segment[:until]))
    return int(segment[index]), segment[index+1:]

def highest_sequence_of_length(number: str, target_length: int):
    sequence = 0
    while target_length > 0:
        digit, number = highest_in_segment(number, target_length)
        sequence = sequence * 10 + digit  # shift digits left and append
        target_length -= 1
    return sequence
```

```python
part1, part2 = 0, 0

for num in data.splitlines():
    part1 += highest_sequence_of_length(num, 2)
    part2 += highest_sequence_of_length(num, 12)

print(f"Part 1: {part1}")
print(f"Part 2: {part2}")
```
