```python
with open("../data/2025/day3.txt") as f:
    data = f.read()
```

```python
data = """987654321111111
811111111111119
234234234234278
818181911112111"""
```

```python
part1 = 0

for line in data.splitlines():
    batteries = list(map(int, list(line)))
    head = batteries.index(max(batteries[:-1]))      # highest battery from the left excluding last
    tail = batteries.index(max(batteries[head+1:]))  # next highest battery from the right of max
    part1 += batteries[head]*10 + batteries[tail]

print(f"Part 1: {part1}")
```
