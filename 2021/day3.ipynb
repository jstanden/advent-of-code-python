```python
with open("../data/2021/day3.txt") as f:
    lines = f.read().splitlines()
```

```python
bits = list(map(lambda line: int(line, 2), lines))
bit_length = len(lines[0])
reverse_powers_of_2 = [pow(2,p) for p in range(bit_length-1,-1,-1)]
```

```python
def most_common_bit_at(numbers: list[int], bit: int) -> int:
    set_bits = [n & bit for n in numbers]
    a, b = len(list(filter(None,set_bits))), len(set_bits)/2
    return (a > b) - (a < b) # comparator (1=on, -1=off, 0=tied)

def reduce_bits(numbers: list[int], is_most=True) -> int:
    for p in reverse_powers_of_2:
        mcb = most_common_bit_at(numbers, p)

        if mcb == 0: mcb = 1 if is_most else 0        # if tied, most=1 least=0
        else: mcb = mcb > 0 if is_most else mcb < 0   # invert

        for i, n in enumerate(numbers):
            if n & p != mcb*p and len(numbers) > 1: numbers[i] = 0  # remove non-matches until last
        numbers = list(filter(None, numbers))
    return numbers[0]
```

```python
most_common_bits = [1 if most_common_bit_at(bits, p) > -1 else 0 for p in reverse_powers_of_2]
gamma = int("".join(map(str,most_common_bits)), 2)  # bit string to binary
epsilon = gamma ^ int("1"*bit_length, 2)     # XOR invert
print(f"Part 1: {gamma*epsilon}")
```

```python
o2 = reduce_bits(list(bits), True)
co2 = reduce_bits(list(bits), False)
print(f"Part 2: {o2*co2}")
```
