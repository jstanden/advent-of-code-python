```python
from itertools import batched, chain
from collections import Counter
import re
```

```python
with open("../data/2021/day4.txt") as f:
    lines = f.read().splitlines()
```

```python
order = [int(n) for n in lines[0].split(',')]
boards = list(batched([s for s in lines[2:] if len(s)], 5))
boards = [[tuple(int(n) for n in re.findall(r'\b(\d+)\b', line)) for line in board] for board in boards]
```

```python
def bingo():
    part1, part2 = None, None
    winners = set()
    for numbers_drawn, number in enumerate(order):
        for board_id, board in enumerate(boards):
            if board_id in winners: continue
            combos = board + list(zip(*board))  # All rows + all columns
            if any([all(combo) for combo in [[c in order[0:numbers_drawn]
                for c in combos[y]] for y in range(len(combos))]]):  # If any full row or column is bingo
                    score = sum(Counter(chain.from_iterable(board)) - Counter(order[0:numbers_drawn]))  # difference
                    if None is part1: part1 = score * order[0:numbers_drawn][-1]
                    else: part2 = score * order[0:numbers_drawn][-1]
                    winners.add(board_id)
    return part1, part2

print("Part 1: {}\nPart 2: {}".format(*bingo()))   # 58838
```
