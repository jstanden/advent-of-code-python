```python
from collections import defaultdict
```

```python
with open("../data/2015/day16.txt") as f:
    lines = f.read().splitlines()
```

```python
grid = defaultdict(None)
border = (-1+1j, 1j, 1+1j, -1, 1, -1-1j, -1j, 1-1j) # moore neighborhood
corners = ()

def animate(g: defaultdict, part2: bool = False):
    new_grid = defaultdict(None)
    for xy, state in g.items():
        if xy in corners and part2:
            new_state = '#'
        else:
            neighbors_on = sum([1 for offset in border if g.get(xy + offset) == '#'])
            if state == '#': new_state = '#' if neighbors_on in (2, 3) else '.'
            else: new_state = '#' if neighbors_on == 3 else '.'
        new_grid[xy] = new_state
    return new_grid
```

```python
# Load the grid
for y, line in enumerate(reversed(lines)):
    if y == 0: corners = ((0+0j), (0+(y_max := len(lines)-1)*1j), ((x_max := len(line)-1)+0j), (x_max+y_max*1j))
    for x, tile in enumerate(list(line)):
        grid[x + y * 1j] = tile
```

```python
part1_grid = grid
for _ in range(100): part1_grid = animate(part1_grid)
print("Part 1: " + str(sum([1 for tile in part1_grid.values() if tile == '#'])))
```

```python
part2_grid = grid
part2_grid.update({xy:'#' for xy in corners})
for _ in range(100): part2_grid = animate(part2_grid, part2=True)
print("Part 2: " + str(sum([1 for tile in part2_grid.values() if tile == '#'])))
```
