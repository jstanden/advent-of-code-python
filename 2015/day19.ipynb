```python
import re
from collections import defaultdict
from random import shuffle
```

```python
with open("../data/2015/day19.txt") as f:
    data = f.read()
```

```python
formulas = defaultdict(set)
formulas_text, starting_molecule = data.split('\n\n')
for line in formulas_text.splitlines():
    k,v = line.split(' => ')
    formulas[k].add(v)
```

```python
part1 = set()
for atom, swaps in formulas.items():
    for match in re.finditer(atom, starting_molecule):
        for swap in swaps:
            part1.add(starting_molecule[:match.start()] + swap + starting_molecule[match.end():])

print(f"Part 1: {len(part1)}")
```

```python
formulas_reversed = {}
for a, replacements in formulas.items():
    for r in replacements:
        if r != 'e': formulas_reversed[r] = a

def reverse_path(target: str, formulas_to_try: dict):
    items = list(formulas_to_try.items())
    shuffle(items)
    formulas_attempt = dict(items)
    path = []

    while True:
        found = False
        for sequence, a in formulas_attempt.items():
            if sequence in target:
                target = target.replace(sequence, a, 1)  # Only replace the first occurrence
                path.append((sequence, a))
                found = True
                break
        if not found:  # give up if we reach a dead end
            return None
        if target == 'e':
            return path

while True:
    if part2 := reverse_path(starting_molecule, formulas_reversed):
        print(f"Part 2: {len(part2)}")
        break
```
